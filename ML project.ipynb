```python
# import packages
import pandas as pd
import numpy as np
import io
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.naive_bayes import MultinomialNB
from google.colab import files
uploaded = files.upload()
```

```python
# import data
spam_df=pd.read_csv(io.BytesIO(uploaded['spam.csv']))
spam_df
```

```python
# inspect data
spam_df.groupby('Category').describe()
```

```python
# creating new column 'spam'
spam_df['spam']=spam_df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
spam_df
```

```python
# create train/test split
# x -> features (content of emails)
# y -> label (1 or 0)
x_train, x_test, y_train, y_test = train_test_split(spam_df.Message, spam_df.spam)
```

```python
# find word count and store data as a matrix
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
```

```python
print(x_train_count.toarray().shape)
```

```python
# train model
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
# pre-testing ham
email_ham = ['Reminder from']
email_ham_count = cv.transform(email_ham)
if(str(model.predict(email_ham_count)) == '[0]'):
  print("ham")
else:
  print("spam")
```

```python
# test model
x_test_count = cv.transform(x_test)
model.score(x_test_count, y_test)*100
```
